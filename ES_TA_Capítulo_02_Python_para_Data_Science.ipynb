##### Indices, Slicing y Transformaciones

```
# Elegir un valor en una matriz
arr_bis = np.array([[1,2,3],
                    [7,1,6],
                    [9,6,3]])
arr_bis[0][0]
```

```
arr_bis[0,0]
```

```
# Elija la submatriz en la matriz
arr_bis[0:2,0:1]
```

```
# Elija una columna o una fila
print(arr_bis[1,:])
print(arr_bis[:,1])
```

```
# M√°ximo
# De la matriz
print(arr_bis.max())

# Por Filas
print(arr_bis.max(axis=1))

# Por Columnas
print(arr_bis.max(axis=0))
```

```
# M√≠nimo
# De la matriz
print(arr_bis.min())

# Por Filas
print(arr_bis.min(axis=1))

# Por Columnas
print(arr_bis.min(axis=0))
```

```
# Media
# De la matriz
print(arr_bis.mean())

# Por Filas
print(arr_bis.mean(axis=1))

# Por Columnas
print(arr_bis.mean(axis=0))
```

```
# Desviaci√≥n T√≠pica
# De la matriz
print(arr_bis.std())

# Por Filas
print(arr_bis.std(axis=1))

# Por Columnas
print(arr_bis.std(axis=0))
```

```
# Otras formas
print(np.mean(arr_bis, axis=0))
print(np.std(arr_bis, axis=0))
```

```
# Logaritmo
np.log(arr_bis)
```

```
# Exponencial
np.exp(arr_bis)
```

```
# Ra√≠z Cuadrada
np.sqrt(arr_bis)
```

```
# Concatenar 
# Par√°metros: tupla de arrays y el eje de la concatenaci√≥n
# Concatenar 1 dimensi√≥n
arr1 = arr_bis[:,1]
arr2 = arr_bis[:,2]

print(f"ARR1 {arr1}")
print(f"ARR2 {arr2}")

print(np.concatenate((arr1, arr2), axis=0))
```

```
# Concatenar 2 dimensi√≥n
arr1 = arr1.reshape(-1,1)
arr2 = arr2.reshape(-1,1)

print(f"ARR1 {arr1}")
print(f"ARR2 {arr2}")

print(np.concatenate((arr1, arr2), axis=0))
```

```
# Concatenar 2 dimensi√≥n
print(np.concatenate((arr1, arr2), axis=1))
```

##### Cleaning and selection

```
# Importar un CSV
assets = pd.read_csv("/content/assets.csv", parse_dates=True, index_col="time")
assets
```

```
# Ordenar
assets = assets.sort_index(ascending=True)
assets
```

```
# Seleccionar una columna
assets = assets[["Close DJI30", "Close CAC40", "Close SP500"]]
assets
```

```
# Eliminar los valores faltantes o NAN
assets = assets.dropna()
assets
```

```
# Reiniciar los √≠ndices
assets.reset_index(drop=True)
```

```
# Rolling 
assets["SMA 15"] = assets[["Close SP500"]].rolling(15).mean()
assets
```

```
# Shift
assets["Close SP500"].shift(1)
```

```
arr = [[1,11],
       [1,10],
       [3,15],
       [3,15],
       [3,10]]

df = pd.DataFrame(arr, columns=["num","value"])

df
```

```
# Agrupar por media
df.groupby(by="num").mean()
```

```
# Agrupar por suma
df.groupby(by="num").sum()
```

```
# Agrupar por desviaci√≥n t√≠pica
df.groupby(by="num").std()
```

##### Representaci√≥n gr√°fica

```
import numpy as np
```

```
# Simulaci√≥n de datos
arr = np.random.randint(0,15,size=(15,))

# Cuadrados del array
square = arr**2

# Productos del array
mut = arr*3

print(arr)
print(square)
print(mut)
```

```
# Representaci√≥n de los datos
plt.plot(arr)
plt.plot(square)
plt.plot(mut)

# Cambiar etiqueta del eje x
plt.xlabel("N√∫meros Aleatorios")

# Cambiar etiqueta del eje y
plt.ylabel("Valores")

# Cambiar etiqueta del t√≠tulo
plt.title("Gr√°fico")

# Colocar una leyenda
plt.legend(["Normales", "Cuadrados", "Productos"])

# Mostrar el gr√°fico
plt.show()
```

##### Scatter

```
# Nube de puntos
plt.scatter(arr, mut)
plt.scatter(arr, square)

# Cambiar etiqueta del eje x
plt.xlabel("Random Numbers")

# Cambiar etiqueta del eje y
plt.ylabel("Values")

# Cambiar etiqueta del t√≠tulo
plt.title("Scatter")

# Colocar una leyenda
plt.legend(["Producto", "Cuadrado"])

# Mostrar el gr√°fico
plt.show()
```
